### train-test split
split by giving importance to 'CHAS' using StratifiedShuffleSplit

````python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2, random_state=42)  # n_splits=1 : number of iteration

for train_index, test_index in split.split(df,df['CHAS']):  # n_splits=1 so only one loop, output is generator so use for loop
  train_set = df.loc[train_index]
  test_set = df.loc[test_index]
````

### model selection - Random forest regressor or DecisionTreeRegressor
By performing 10 fold cross validation

````python
X_train = train_set.drop(columns=['MEDV'])
y_train = train_set['MEDV']

X_test = test_set.drop(columns=['MEDV'])
y_test = test_set['MEDV']
````

````python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

model = RandomForestRegressor()
cross_result = cross_val_score(model, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
rmse = np.sqrt(-cross_result)
print('rmse: ',rmse)
print('mean: ',np.mean(rmse))
````
